## **I.I-Model**

For each message, the emitter is given several subcarriers for a few time
slots, this builds a subchannel. In principle, the number of allocated frequency
bins can vary from one slot to the other for a given user, however, we here
consider that this number is fixed during the ON period, hence the rectangular
form of the TXs. Let $N(x)$ be the number of subcarriers needed by a user
located at $x$, who wants to communicate at a rate $C$:

$$
N(x) = \left\lceil \frac{C}{W \log_2 \left( 1 + K \|x\|^{-\gamma} \right)} \right\rceil \quad \text{if} \quad K \|x\|^{-\gamma} > \text{SNR}_{\min},
\ $$

where $\text{SNR}_{\min}$ is the minimum signal over noise ratio necessary to ensure the
feasibility of the communication. Moreover, $\lceil z \rceil$ is the least integer greater
than the real $z$, $W$ is the bandwidth of each frequency bin and $K$ is a constant
which takes into account emitting power, fading, and shadowing.

Note that the number of slots per user is bounded by 
$$ N_{\max} = \left\lceil \frac{C}{W \log_2 \left( 1 + \text{SNR}_{\min} \right)} \right\rceil. $$

# **II-Questions**

We consider that users are represented by a Poisson process of intensity
measure $\lambda \, dx$ and that there is a percentage $p$ of active users in a time
slot. We want to estimate for one time slot the outage probability, i.e., the
probability that the cumulative number of required subchannels by the set
of active customers is greater than the number of available resources $N_{\text{avail}}$,
which is 25 on Figure 1.

The access point is located at $(0, 0)$. The observation window is a circle
of radius $R$.


##### **Question 1**: Show that the process of active customers is a Poisson process. Give its intensity measure.

Notons $P$ le processus de de Poisson qui représente les utilisateurs:

Notons $ P_{p} $ processus qui compte le nombre de client actif. On remarque que ce processus est un amincissement de $P$ avec pour chaque point de $P_t$ (Processus $P$ à l'instant $t$) une probabilité de $p$ qu'il soit actif. 

Donc : $$ P_p = p \circ P \sim P.P.P(p\lambda dx)$$

avec P.P.P : Processus de Poisson Ponctuel.

##### **Question 3**: For $k = 1, · · · , N_{max}$, characterize geometrically the subset of the cell in which all customers require $k$ subcarriers.

$$ \implies  \|x\| \in \left](\frac{K}{2^{\frac{C}{W(k-1)}}-1})^\frac{1}{\gamma}, (\frac{K}{2^{\frac{C}{Wk}}-1})^\frac{1}{\gamma} \right]  $$

##### **Question 7**: Compute $ E \left[ e^{\theta X_\mu} \right] $ when $ X_\mu $ is a Poisson random variable of parameter $ \mu $ and find $ \min_{\theta > 0} e^{-K_{\mu}\theta} E \left[ e^{\theta X_\mu} \right] $.

Premièrement:

$$ E \left[ e^{\theta X_\mu} \right] = \sum_{k \geq 0} e^{\theta k} \mathbb{P}( X_\mu = k) 
=  \sum_{k \geq 0} e^{\theta k} \frac{\mu^k}{k!}e^{-\mu} = \sum_{k \geq 0} \frac{(e^{\theta}\mu)^k}{k!} e^{-\mu}
=e^{\mu( e^{\theta} - 1)} $$

Deuxièmement :

$$  \min_{\theta > 0} e^{-K_{\mu}\theta} E \left[ e^{\theta X_\mu} \right] 
=  \min_{\theta > 0}  e^{ \mu(e^{\theta} - 1) - K_{\mu}\theta} $$

Posons $ f(\theta) = \mu(e^{\theta} - 1) - K_{\mu}\theta $, on a $f'(\theta) = \mu e^{\theta} -K_{\mu} $. 

$f$ est décroissant sur $] -\infty; ln(\frac{K_{\mu}}{\mu})] $ et croissant sur $[ ln(\frac{K_{\mu}}{\mu}) ; +\infty[ $.

Finalement :

$$ \min_{\theta > 0} e^{-K_{\mu}\theta} E \left[ e^{\theta X_\mu} \right] = e^{K_{\mu}- \mu -K_{\mu}ln(\frac{K_{\mu}}{\mu}) )}$$ 

##### **Question 8**: Using the fact that $ P(X_\mu \geq K\mu) \leq e^{-K\mu \theta} E \left[ e^{\theta X_\mu} \right] $, find $ K_\mu $ such that $ P(X_\mu \geq K_\mu) \leq 10^{-4} $.


En utilisant la question précédente on obtient :

$$ P(X_\mu \geq K\mu) \leq \min_{\theta > 0} e^{-K\mu\theta} E \left[ e^{\theta X_\mu} \right] \leq e^{-K\mu \theta} E \left[ e^{\theta X_\mu} \right]       $$

$$ \implies P(X_\mu \geq K\mu) \leq \min_{\theta > 0} e^{-K\mu\theta} E \left[ e^{\theta X_\mu} \right] 
= e^{K_{\mu}- \mu -K_{\mu}ln(\frac{K_{\mu}}{\mu}) )}$$

et $ e^{K_{\mu}- \mu -K_{\mu}ln(\frac{K_{\mu}}{\mu}) )} \leq 10^{-4} \Leftrightarrow K_{\mu}(1 + ln(\mu)-ln(K_{\mu})) \leq -4ln(10) + \mu:= \alpha $

On pose $ g(K_{\mu}) =  K_{\mu}(1 + ln(\mu)-ln(K_{\mu})) - \alpha$, on a $ g'(K) = 1 -(ln(K) +1) = -ln(K) $

Donc $ g $ est croissant sur $]0; \mu]$  et décroissante sur $ [\mu; +\infty[$ , en particulier g est bijective sur chacun de ces 2 intervalles. Cela nous donne l'existence d'au moins 2 valeurs.

Condons une fonction qui donne la valeur de $K$ sur $ [\mu; +\infty[$.

```python
#Codons une fonction qui donne une valeur de K qui marche
import numpy as np
from scipy.optimize import brentq
def g(x,mu):
    alpha = -4*np.log(10) + mu
    return x*(1 + np.log(mu) - np.log(x))- alpha

#On cherche la valeur de K qui donne g(K, mu) = 0
#valeur test de mu
mu = 20

zero_x = brentq(lambda x: g(x, mu), mu, 10000000)
print("K = ",zero_x)
```

##### **Question 9**:

```python
#Import
import numpy as np
import scipy
from scipy.optimize import brentq
```

```python
#Valeurs numériques
C = 200
W = 250
K = 10**6
gamma = 2.8
R = 300
lamb= 0.01
SNR = 0.1
p = 0.01
S_min = 30
S_max = 100
```

##### Calcul $N_{max}$

```python
N_max = int(C / (W*np.log2( 1 + SNR))) + 1
print('N_max =', N_max)
```

##### Compute the mean value of the r.v. $\zeta_k$ for k = 1,...,$N_{max}$ 

D'après la question 6 on a : $\zeta_k \sim \mathcal{P}(p\lambda q_k)$. Donc $p\lambda q_k$ est la moyenne de $\zeta_k$.

Avec $ q_k = \pi ((s_k^{(2)})^2 - (s_k^{(1)})^2 ),
s_k^{(1)} = (\frac{K}{2^{\frac{C}{W(k-1)}}-1})^\frac{1}{\gamma}$ et $ s_k^{(2)} = (\frac{K}{2^{\frac{C}{Wk}}-1})^\frac{1}{\gamma}  $.

```python
def mean_zeta(k):
    ''' Attention, on traite le cas k=0 à part pour éviter la divison par 0
    '''
    if k > 1 :
        s_1 = ( K / ( 2**( C/(W*(k-1) )) - 1 )) ** (1/gamma)
        s_2 = ( K / ( 2**( C/(W*k) ) - 1 )) ** (1/gamma)
        q_k = np.pi * (s_2**2 - s_1**2)
    else :
        s_2 = ( K / ( 2**( C/(W*k) ) - 1 )) ** (1/gamma)
        q_k = np.pi * (s_2**2)
    return lamb*p*q_k 

for k in range(1, N_max+1):
    print("La valeur moyenne de zeta_"+ str(k)+ " est " + str(mean_zeta(k)))
```

##### Compute the size of the vector which represents the distribution of $\zeta_k$ for k = 1,...,$N_{max}$ and take K to be the supremum of these dimensions

```python
def size(k):
    mu = mean_zeta(k)
    zero_x = brentq(lambda x: g(x, mu), mu, 1000000)
    return int(zero_x)+1

for i in range(1, N_max+1):
    print("La taille du vecteur de zeta_"+ str(i)+ " est " + str(size(i)))
```

##### Construct the vectors representing the distribution of the $\zeta_ks$.

```python
def vector_zeta(k):
    mu = mean_zeta(k)
    K = size(k)
    return np.array([np.exp(-mu)* (mu**k / scipy.special.factorial(k)) for k in range(K+1)])

#plot des vecteurs 6 vecteurs
import matplotlib.pyplot as plt

for i in range(1, 7):  
    plt.scatter(np.arange(size(i)+1), vector_zeta(i), label = 'zeta_'+str(i))
    plt.legend()
    plt.xlabel('k')
    plt.ylabel('zeta_k')
    plt.title('Vecteurs zeta'+str(i))
    plt.show()
```

##### Construct the vectors representing the distribution of the $k\zeta_ks$.

```python
def vector_kzeta(k):
    mu = mean_zeta(k)
    K = size(k)
    res = np.zeros(k*(K+1))
    for i in range(K+1):
        res[i*k] = np.exp(-mu)* (mu**i / scipy.special.factorial(i))
    return res

#plot des vecteurs 6 vecteurs
import matplotlib.pyplot as plt

for i in range(1, 7):  
    plt.scatter(np.arange(i*(size(i)+1)), vector_kzeta(i), label = 'kzeta_'+str(i))
    plt.legend()
    plt.xlabel('k')
    plt.ylabel('kzeta_k')
    plt.title('Vecteurs zeta'+str(i))
    plt.show()
```

##### Using the numpy.convolve, build the vector representing the distribution of $F$.

Rappel : Soit $Z = X + Y $ tel que $ X \perp\!\!\!\perp Y$ alors $ \mathbb{P}_Z = \mathbb{P}_X* \mathbb{P}_Y$ 

```python
def vectorF():
    res = vector_kzeta(1)
    for i in range(2, N_max+1):
        res = np.convolve(res, vector_kzeta(i))
    return res

resF = vectorF()
```

##### **Question 10** : Compute $S$ such that the outage probability is smaller than 0.01

```python
somme = 0
prob = 0.01
ind = len(resF)-1
while somme < prob and ind >= 0:
    somme += resF[ind]
    ind -= 1
ind += 1

print("On a S =", ind,", tel que P(F>S)< 0.01")
print("Verfication : P(F >",ind,") =", resF[0:ind+1].sum(), "et P(F >",ind-1,") =" , resF[0:ind].sum() )
```
